# 本題要實作一個紀錄頻率的函數 (\#Design, \#HashMap )
原題目連結: https://leetcode.com/problems/frequency-tracker/description/  

Design a data structure that keeps track of the values in it and answers some queries regarding their frequencies.  
Implement the FrequencyTracker class.  

FrequencyTracker(): Initializes the FrequencyTracker object with an empty array initially.  
void add(int number): Adds number to the data structure.  
void deleteOne(int number): Deletes one occurrence of number from the data structure. The data structure may not contain number, and in this case nothing is deleted.  
bool hasFrequency(int frequency): Returns true if there is a number in the data structure that occurs frequency number of times, otherwise, it returns false.  

範例:  
Ex1:  
Input  
["FrequencyTracker", "add", "add", "hasFrequency"]  
[[], [3], [3], [2]]  
Output  
[null, null, null, true]  

Explanation  
FrequencyTracker frequencyTracker = new FrequencyTracker();  
frequencyTracker.add(3); // The data structure now contains [3]  
frequencyTracker.add(3); // The data structure now contains [3, 3]  
frequencyTracker.hasFrequency(2); // Returns true, because 3 occurs twice  

Ex2:  
Input  
["FrequencyTracker", "add", "deleteOne", "hasFrequency"]  
[[], [1], [1], [1]]  
Output  
[null, null, null, false]  

Explanation  
FrequencyTracker frequencyTracker = new FrequencyTracker();  
frequencyTracker.add(1); // The data structure now contains [1]  
frequencyTracker.deleteOne(1); // The data structure becomes empty []  
frequencyTracker.hasFrequency(1); // Returns false, because the data structure is empty  

Ex3:  
Input  
["FrequencyTracker", "hasFrequency", "add", "hasFrequency"]  
[[], [2], [3], [1]]  
Output  
[null, false, null, true]  

Explanation  
FrequencyTracker frequencyTracker = new FrequencyTracker();  
frequencyTracker.hasFrequency(2); // Returns false, because the data structure is empty  
frequencyTracker.add(3); // The data structure now contains [3]  
frequencyTracker.hasFrequency(1); // Returns true, because 3 occurs once

* 解題想法:  
首先用一個dict用來存放每個數字出現的次數，另外用一個dict來存放每個頻率出現的次數，接著開始實作題目要求的功能:  
- add(): 首先檢查數值是否有出現過，如果沒有的話則將這個數字放入dict中，如果有的話則更新dict中出現的次數，接著更新出現的頻率  
- deleteOne(): 先檢查這個數值是否有出現，如果有出現的話，則先將該數值出現次數減一，如果次數歸零則刪除這個數字，另外也將出現頻率的dict進行更新  
- hasFrequency(): 如果指定的值有出現在頻率的dict中則回傳True，否則回傳False  

```python
class FrequencyTracker:

    def __init__(self):
        self.data = {}
        self.freq = {}
    def add(self, number: int) -> None:
        if number in self.data.keys():
            if self.data[number] in self.freq.keys():
                self.freq[self.data[number]] -= 1
                if self.freq[self.data[number]] == 0:
                    self.freq.pop(self.data[number])
            self.data[number] += 1
        else:
            self.data[number] = 1
        if self.data[number] in self.freq.keys():
            self.freq[self.data[number]] += 1
        else:
            self.freq[self.data[number]] = 1

    def deleteOne(self, number: int) -> None:
        if number in self.data.keys():
            if self.data[number] == 1:
                self.data.pop(number)
                self.freq[1] -= 1
                if self.freq[1] == 0:
                    self.freq.pop(1)
            else:
                self.freq[self.data[number]] -= 1
                if self.freq[self.data[number]] == 0:
                    self.freq.pop(self.data[number])
                if self.data[number] -1 in self.freq.keys():
                    self.freq[self.data[number]-1] += 1
                else:
                    self.freq[self.data[number]-1] = 1
                self.data[number] -= 1

    def hasFrequency(self, frequency: int) -> bool:
        if frequency in self.freq.keys():
            return True
        return False


# Your FrequencyTracker object will be instantiated and called as such:
# obj = FrequencyTracker()
# obj.add(number)
# obj.deleteOne(number)
# param_3 = obj.hasFrequency(frequency)
```
