# 本題要找出三個數字總和能為三的倍數的最大值為多少 (\#Array, \#Math, \#HashMap )
原題目連結: https://leetcode.com/problems/maximum-sum-of-three-numbers-divisible-by-three/description/  

You are given an integer array nums.  
Your task is to choose exactly three integers from nums such that their sum is divisible by three.  
Return the maximum possible sum of such a triplet. If no such triplet exists, return 0.  

範例:  
Ex1:  
Input: nums = [4,2,3,1]  
Output: 9  
Explanation:  
The valid triplets whose sum is divisible by 3 are:  
- (4, 2, 3) with a sum of 4 + 2 + 3 = 9.  
- (2, 3, 1) with a sum of 2 + 3 + 1 = 6.  

Thus, the answer is 9.  

Ex2:  
Input: nums = [2,1,5]  
Output: 0  
Explanation:  
No triplet forms a sum divisible by 3, so the answer is 0.

* 解題想法:  
首先用一個dict來存放餘數分別為0，1，2的數值，接著檢查以下的幾種組合:
- 三個數分別為餘0，1，2的組合
- 三個數都為餘1的情形  
- 三個數都為餘2的情形  
- 三個數都為餘0的情形  
最後找出最大值就是答案

```python
class Solution:
    def maximumSum(self, nums: List[int]) -> int:
        res = 0
        count = {}
        nums.sort()
        for n in nums:
            d = n % 3
            if d not in count.keys():
                count[d] = []
            count[d].append(n)
        if len(count) == 3:
            res = count[0][-1] + count[1][-1] +count[2][-1]
        if 0 in count.keys() and len(count[0]) >= 3:
            res = max(res, count[0][-1] + count[0][-2] + count[0][-3])
        if 1 in count.keys() and len(count[1]) >= 3:
            res = max(res, count[1][-1] + count[1][-2] + count[1][-3])
        if 2 in count.keys() and len(count[2]) >= 3:
            res = max(res, count[2][-1] + count[2][-2] + count[2][-3])
        return res
```
