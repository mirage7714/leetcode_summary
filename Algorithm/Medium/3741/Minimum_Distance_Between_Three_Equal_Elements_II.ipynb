# 本題要找出相同三個數字的最短距離為何 (\#Array, \#HashMap ) (解法與No.3740相同 ) 
原題目連結: https://leetcode.com/problems/minimum-distance-between-three-equal-elements-ii/description/  

You are given an integer array nums.  
A tuple (i, j, k) of 3 distinct indices is good if nums[i] == nums[j] == nums[k].  
The distance of a good tuple is abs(i - j) + abs(j - k) + abs(k - i), where abs(x) denotes the absolute value of x.  
Return an integer denoting the minimum possible distance of a good tuple. If no good tuples exist, return -1.

範例:  
Ex1:  
Input: nums = [1,2,1,1,3]  
Output: 6  
Explanation:  
The minimum distance is achieved by the good tuple (0, 2, 3).  
(0, 2, 3) is a good tuple because nums[0] == nums[2] == nums[3] == 1. Its distance is abs(0 - 2) + abs(2 - 3) + abs(3 - 0) = 2 + 1 + 3 = 6.  

Ex2:  
Input: nums = [1,1,2,3,2,1,2]  
Output: 8  
Explanation:  
The minimum distance is achieved by the good tuple (2, 4, 6).  
(2, 4, 6) is a good tuple because nums[2] == nums[4] == nums[6] == 2. Its distance is abs(2 - 4) + abs(4 - 6) + abs(6 - 2) = 2 + 2 + 4 = 8.  

Ex3:  
Input: nums = [1]  
Output: -1  
Explanation:  
There are no good tuples. Therefore, the answer is -1.

* 解題想法:    
首先將所有數字出現的頻率放入dict中，然後接著檢查每個有出現三個以上的數字並找出他們三點之間的距離後取最小值就是答案

```python
class Solution:
    def minimumDistance(self, nums: List[int]) -> int:
        if len(nums) < 3:
            return -1
        res = -1
        count = {}
        for i in range(len(nums)):
            n = nums[i]
            if n not in count.keys():
                count[n] = []
            count[n].append(i)
        for val in count.values():
            if len(val) > 2:
                for i in range(len(val) -2):
                    t = abs(val[i] - val[i + 1]) + abs(val[i] - val[i+2]) + abs(val[i+1] - val[i+2])
                    if res == -1:
                        res = t
                    else:
                        res = min(res, t)
        return res
```
