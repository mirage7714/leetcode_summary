# 本題要找出那些建築物的四面都被包圍了 (\#Matrix, \#HashMap )
原題目連結: https://leetcode.com/problems/count-covered-buildings/description/?envType=daily-question&envId=2025-12-11  

You are given a positive integer n, representing an n x n city. You are also given a 2D grid buildings, where buildings[i] = [x, y] denotes a unique building located at coordinates [x, y].  
A building is covered if there is at least one building in all four directions: left, right, above, and below.  
Return the number of covered buildings.  

範例:  
Ex1:  
Input: n = 3, buildings = [[1,2],[2,2],[3,2],[2,1],[2,3]]  
Output: 1  
Explanation:  
Only building [2,2] is covered as it has at least one building:  
- above ([1,2])  
- below ([3,2])  
- left ([2,1])  
- right ([2,3])  

Thus, the count of covered buildings is 1.  

Ex2:  
Input: n = 3, buildings = [[1,1],[1,2],[2,1],[2,2]]  
Output: 0  
Explanation:  
No building has at least one building in all four directions.  

Ex3:  
Input: n = 5, buildings = [[1,3],[3,2],[3,3],[3,5],[5,3]]  
Output: 1  
Explanation:  
Only building [3,3] is covered as it has at least one building:  
- above ([1,3])  
- below ([5,3])  
- left ([3,2])  
- right ([3,5])  

Thus, the count of covered buildings is 1.  

* 解題想法:  
先將X軸與Y軸的座標放入dict中，並分別放入Y軸與X軸的座標，接著從X軸進行檢查，去掉頭尾兩個值之後開始檢查中間的值是否有出現在Y軸的dict中，如果有的話則檢查該X軸是否為左右兩端，如果不是則將結果加1，最後回傳統計結果  

```python
class Solution:
    def countCoveredBuildings(self, n: int, buildings: List[List[int]]) -> int:
        res = 0
        x_count = {}
        y_count = {}
        for b in buildings:
            if b[0] not in x_count.keys():
                x_count[b[0]] = []
            x_count[b[0]].append(b[1])
            if b[1] not in y_count.keys():
                y_count[b[1]] = []
            y_count[b[1]].append(b[0])
        for key, val in x_count.items():
            val.sort()
            for i in range(1, len(val)-1):
                v = val[i]
                if v in y_count.keys():
                    y_count[v].sort()
                    if key != y_count[v][0] and key != y_count[v][-1]:
                        res += 1
        return res
```
