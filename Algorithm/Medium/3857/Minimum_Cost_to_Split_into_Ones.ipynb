# 本題要找出最小將n分割成1的成本為多少 (\#Math )
原題目連結: https://leetcode.com/problems/minimum-cost-to-split-into-ones/description/  

You are given an integer `n`.  
In one operation, you may split an integer `x` into two positive integers `a` and `b` such that `a + b = x`.  
The cost of this operation is `a * b`.   
Return an integer denoting the minimum total cost required to split the integer `n` into n ones.

範例:  
Ex1:  
Input: n = 3  
Output: 3  
Explanation:  
One optimal set of operations is:
|x | a | b | a + b | a * b | Cost |
| --- | --- | --- | --- | --- | --- | 
|3	|1	|2	|3	|2	|2|
|2	|1	|1	|2	|1	|1|

Thus, the minimum total cost is `2 + 1 = 3`.

Ex2:  
Input: n = 4  
Output: 6  
Explanation:  
One optimal set of operations is:  
|x	|a	|b	|a + b	|a * b|	Cost|
| --- | --- | --- | --- | --- | --- |
|4	|2	|2	|4	|4	|4|
|2	|1	|1	|2	|1	|1|
|2	|1	|1	|2	|1	|1|

Thus, the minimum total cost is `4 + 1 + 1 = 6`.

* 解題想法:  
首先用一個stack來存放目前切分後的數值，並開始取值，當該數值大於1的情況，進行兩等分之後算出成本後重新放回stack中，最後回傳總成本就是答案

```python
class Solution:
    def minCost(self, n: int) -> int:
        res = 0
        stack = [n]
        while stack:
            l = stack.pop()
            if l > 1:  
                j = l // 2
                if l % 2 == 0:
                    res += j * j
                    if j != 1:
                        stack.append(j)
                        stack.append(j)
                else:
                    res += j * (j + 1)
                    if j != 1:
                        stack.append(j)
                    stack.append(j + 1)
        return res
```
