# 本題要找出符合條件的matrix共有多少 (\#Matrix, \#Array )
原題目連結: https://leetcode.com/problems/magic-squares-in-grid/description/?envType=daily-question&envId=2025-12-30  

A 3 x 3 magic square is a 3 x 3 grid filled with distinct numbers from 1 to 9 such that each row, column, and both diagonals all have the same sum.  
Given a row x col grid of integers, how many 3 x 3 magic square subgrids are there?  
Note: while a magic square can only contain numbers from 1 to 9, grid may contain numbers up to 15.  

範例:  
Ex1:  
Input: grid = [[4,3,8,4],[9,5,1,9],[2,7,6,2]]  
Output: 1  
Explanation:   
The following subgrid is a 3 x 3 magic square:   
[[4,3,8],[9,5,1],[2,7,6]]   
while this one is not:  
[[3,8,4],[5,1,9],[7,6,2]]  
In total, there is only one magic square inside the given grid.  

Ex2:  
Input: grid = [[8]]  
Output: 0  

* 解題想法:  
首先建立一個函數來計算每個row與column還有對角的總和，並判斷是不是符合條件，接著找出所有3x3的matrix後進行判斷，最後回傳結果就是答案  

```python
class Solution:
    def numMagicSquaresInside(self, grid: List[List[int]]) -> int:
        
        def verify_matrix(m):
            nums = set()
            count = [0] * 8
            for i in range(3):
                for j in range(3):
                    if m[i][j] in nums:
                        return False
                    if m[i][j] > 9 or m[i][j] < 1:
                        return False
                    nums.add(m[i][j])
                    count[j] += m[i][j]
                    count[i+3] += m[i][j]
                    if (i == 0 and j == 0) or (i == 1 and j == 1) or (i == 2 and j == 2):
                        count[6] += m[i][j]
                    if (i == 0 and j == 2) or (i == 1 and j == 1) or (i == 2 and j == 0):
                        count[7] += m[i][j]
            return max(count) == min(count)

        res = 0
        if len(grid) > 2 and len(grid[0]) > 2:
            for j in range(len(grid[0])-2):
                for i in range(len(grid)-2):
                    s = []
                    s.append(grid[i][j:j+3])
                    s.append(grid[i+1][j:j+3])
                    s.append(grid[i+2][j:j+3])
                    if verify_matrix(s):
                        res += 1
        return res
```
