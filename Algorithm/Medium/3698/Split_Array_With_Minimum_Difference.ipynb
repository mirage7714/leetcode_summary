# 本題要將list依條件進行分割後找出最小的差值為何 (\#Array )
原題目連結：https://leetcode.com/problems/split-array-with-minimum-difference/description/  

You are given an integer array nums.  
Split the array into exactly two subarrays, left and right, such that left is strictly increasing and right is strictly decreasing.  
Return the minimum possible absolute difference between the sums of left and right. If no valid split exists, return -1.  

範例：  
Ex1:  
Input: nums = [1,3,2]  
Output: 2  
Explanation:   
|i|left | right | Validity | left sum | right sum | Absolute difference |
|---|---|---|---|--- | --- | --- | 
|0 | [1] | [3, 2] | Yes | 1 | 5 | \|1 - 5\| = 4 |   
|1 | [1, 3] | [2] | Yes | 4 | 2 | \|4 - 2\| = 2 |   

Thus, the minimum absolute difference is 2.   

Ex2:  
Input: nums = [1,2,4,3]   
Output: 4  
Explanation:   

|i |left | right | Validity | left sum | right sum | Absolute difference |
| ---- | ---- | ---- | --- | --- | --- | --- |
|0 | [1]  | [2, 4, 3] | No | 1 | 9 | - |
|1 |[1, 2] | [4, 3] | Yes | 3 | 7 | \|3 - 7\| = 4 | 
|2 | [1, 2, 4] |[3] | Yes | 7 | 3 | \|7 - 3\| = 4 | 

Thus, the minimum absolute difference is 4.  

Ex3:  
Input: nums = [3,1,2]  
Output: -1  
Explanation:  
No valid split exists, so the answer is -1.  

* 解題想法：  
首先檢查list中是否有同時滿足嚴格遞增與嚴格遞減的條件：  
- 如果list長度為2，則一定符合條件，結果就是兩個數值相減的絕對值   

接著用迴圈開始檢查並將左右兩個list找出來，先從左邊的遞增開始尋找，如果遇到不符合嚴格遞增，則開始放入嚴格遞減的list中，如果遇到不符合嚴格遞減的條件則表示無法符合條件所以回傳-1  
判斷總和時：  
- 如果遇到list都是嚴格遞增的情況，則兩邊差值最小的情形為前面所有數值的總和減去最後一個數值  
- 如果中間有兩數相同，則結果為兩邊總和相減  
- 如果中間沒有相同數值，且符合題目要求條件，則最小值為左邊減去右邊或是左邊減去右邊再減去兩倍的中間值  

```python
class Solution:
    def splitArray(self, nums: List[int]) -> int:
        left = []
        right = []
        p = [0,0]
        b = 0 
        prev = 0
        is_equal = False
        is_right = False
        if len(nums) == 2:
            return abs(nums[0] - nums[1])
        for n in nums:
            if not left:
                left.append(n)
                p[0] += n
            else:
                if left[-1] < n and not is_right:
                    left.append(n)
                    p[0] += n
                else:
                    if not right:
                        is_equal = left[-1] == n
                        right.append(n)
                        p[1] += n
                        b = prev
                        is_right = True
                    elif right and right[-1] > n:
                        right.append(n)
                        p[1] += n
                    else:
                        return -1
            prev = n
        if is_equal:
            return abs(p[0] - p[1])
        elif p[1] == 0:
            return abs(p[0] - 2 * nums[-1])
        return min(abs(p[0]-p[1]), abs(p[0] - p[1] - 2 * b))
```
