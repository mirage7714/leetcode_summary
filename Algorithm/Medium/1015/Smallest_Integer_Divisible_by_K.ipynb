# 本題要找出最小可以被k整除的整數長度為多少 (\#Math )
原題目連結: https://leetcode.com/problems/smallest-integer-divisible-by-k/description/?envType=daily-question&envId=2025-11-25  

Given a positive integer k, you need to find the length of the smallest positive integer n such that n is divisible by k, and n only contains the digit 1.  
Return the length of n. If there is no such n, return -1.  
Note: n may not fit in a 64-bit signed integer.  

範例:  
Ex1:  
Input: k = 1  
Output: 1  
Explanation: The smallest answer is n = 1, which has length 1.  

Ex2:  
Input: k = 2  
Output: -1  
Explanation: There is no such positive integer n divisible by 2.  

Ex3:  
Input: k = 3  
Output: 3  
Explanation: The smallest answer is n = 111, which has length 3.

* 解題想法:  
首先檢查如果k為偶數的話，是不可能出現的，另外如果k為5的倍數的話也是不可能出現，所以先過濾掉這兩種情境，接下來用一個while迴圈進行檢查，如果當餘數是沒有出現過的情況，則將原本的數字乘10之後加上1，並將餘數放入暫存中，直到出現重複的餘數或是整除則中斷迴圈，最後如果是可以整除的情況則回傳整數的長度就是答案  

```python
class Solution:
    def smallestRepunitDivByK(self, k: int) -> int:
        if k % 2 == 0:
            return -1
        elif k % 5 == 0:
            return -1
        else:
            temp = set()
            s = 1
            r = 1
            while s % k != 0:
                if s % k not in temp:
                    temp.add(s % k)
                else:
                    return -1
                    break
                s = s * 10 + 1
                r += 1
            return r
```
