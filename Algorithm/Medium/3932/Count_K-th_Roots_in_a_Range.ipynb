# 本題要找出指定區間內有多少值的k次方 (\#Math )
原題目連結: https://leetcode.com/problems/count-k-th-roots-in-a-range/description/  

You are given three integers l, r, and k.  
An integer y is said to be a perfect kth power if there exists an integer x such that y = x<sup>k</sup>.  
Return the number of integers y in the range `[l, r]` (inclusive) that are perfect k<sup>th</sup> powers.  

範例:  
Ex1:  
Input: l = 1, r = 9, k = 3  
Output: 2  
Explanation:  
The perfect cubes in the range [1, 9] are:  
- 1 = 1<sup>3</sup>  
- 8 = 2<sup>3</sup>

Hence, the answer is 2.  

Ex2:  
Input: l = 8, r = 30, k = 2  
Output: 3  
Explanation:  
The perfect squares in the range [8, 30] are:
- 9 = 3<sup>2</sup>
- 16 = 4<sup>2</sup>
- 25 = 5<sup>2</sup>

Hence, the answer is 3.

* 解題想法:  
首先找出兩個邊界值分別是多少的k次方，接著用迴圈判斷有多少值滿足指定的條件後就是答案，另外如果k為1的情況，中間滿足的值就是兩端相減加一  

```python
class Solution:
    def countKthRoots(self, l: int, r: int, k: int) -> int:
        res = 0
        if k == 1:
            res = r - l + 1
        else: 
            left = floor(pow(l, 1 / k))
            right = ceil(pow(r, 1/ k))
            for i in range(left, right+1):
                p = pow(i, k)
                if p >= l and p <= r:
                    res += 1
        return res
```
