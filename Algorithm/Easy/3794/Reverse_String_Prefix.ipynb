# 本題要找出翻轉長度為k的prefix後的字串為何 (\#String )
原題目連結: https://leetcode.com/problems/reverse-string-prefix/description/  

You are given a string s and an integer k.  
Reverse the first k characters of s and return the resulting string.  

範例:  
Ex1:  
Input: s = "abcd", k = 2  
Output: "bacd"  
Explanation:  
The first k = 2 characters "ab" are reversed to "ba". The final resulting string is "bacd".  

Ex2:  
Input: s = "xyz", k = 3  
Output: "zyx"  
Explanation:  
The first k = 3 characters "xyz" are reversed to "zyx". The final resulting string is "zyx".  

Ex3:  
Input: s = "hey", k = 1  
Output: "hey"  
Explanation:  
The first k = 1 character "h" remains unchanged on reversal. The final resulting string is "hey".

* 解題想法:  
用一個for迴圈，當目前的值小於k的時候用倒轉的方式加到結果，如果大於等於k的情況則加到最後，就可以得到要求的字串  

```python
class Solution:
    def reversePrefix(self, s: str, k: int) -> str:
        res = ''
        for i in range(len(s)):
            if i < k:
                res = s[i] + res
            else:
                res += s[i]
        return res
```
