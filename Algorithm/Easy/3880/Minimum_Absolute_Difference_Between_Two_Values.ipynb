# 本題要找出最小的距離為多少 (\#Array )
原題目連結: https://leetcode.com/problems/minimum-absolute-difference-between-two-values/description/  

You are given an integer array nums consisting only of 0, 1, and 2.  
A pair of indices `(i, j)` is called valid if `nums[i] == 1` and `nums[j] == 2`.  
Return the minimum absolute difference between i and j among all valid pairs. If no valid pair exists, return `-1`.  
The absolute difference between indices i and j is defined as `abs(i - j)`.  

範例:  
Ex1:  
Input: nums = [1,0,0,2,0,1]  
Output: 2  
Explanation:  
The valid pairs are:  
- (0, 3) which has absolute difference of abs(0 - 3) = 3.
- (5, 3) which has absolute difference of abs(5 - 3) = 2.

Thus, the answer is 2.

Ex2:  
Input: nums = [1,0,1,0]  
Output: -1  
Explanation:  
There are no valid pairs in the array, thus the answer is -1.

* 解題想法:  
首先找出1與2的index，接著檢查與前一個指定的數值距離為多少，如果小於目前距離則更新，最後回傳距離就是答案

```python
class Solution:
    def minAbsoluteDifference(self, nums: list[int]) -> int:
        count = {}
        res = -1
        for i in range(len(nums)):
            if nums[i] == 1:
                if 1 not in count.keys():
                    count[1] = []
                count[1].append(i)
                if 2 in count.keys():
                    if res == -1:
                        res = i - count[2][-1]
                    else:
                        res = min(res, i - count[2][-1])
            elif nums[i] == 2:
                if 2 not in count.keys():
                    count[2] = []
                count[2].append(i)
                if 1 in count.keys():
                    if res == -1:
                        res = i - count[1][-1]
                    else:
                        res = min(res, i-count[1][-1])
        return res
```
