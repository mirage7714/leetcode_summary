# 本題要找出第一個只出現一次的偶數 (\#Array, \#HashMap )
原題目連結: https://leetcode.com/problems/first-unique-even-element/description/  

You are given an integer array nums.  
Return an integer denoting the first even integer (earliest by array index) that appears exactly once in nums. If no such integer exists, return -1.  
An integer x is considered even if it is divisible by 2.  

範例:  
Ex1:  
Input: nums = [3,4,2,5,4,6]    
Output: 2    
Explanation:  
Both 2 and 6 are even and they appear exactly once. Since 2 occurs first in the array, the answer is 2.  

Ex2:  
Input: nums = [4,4]  
Output: -1  
Explanation:  
No even integer appears exactly once, so return -1.  

* 解題想法:  
首先用一個dict來找出所有出現過的偶數，接著找出第一個出現且只有一次的偶數之後就是答案，如果沒有則回傳-1

```python
class Solution:
    def firstUniqueEven(self, nums: list[int]) -> int:
        res = -1
        i = inf
        count = {}
        for i in range(len(nums)):
            n = nums[i]
            if n % 2 == 0: 
                if n not in count.keys():
                    count[n] = []
                count[n].append(i)
        for key, val in count.items():
            if len(val) == 1:
                if val[0] < i or res == -1:
                    i = val[0]
                    res = key
        return res
```
