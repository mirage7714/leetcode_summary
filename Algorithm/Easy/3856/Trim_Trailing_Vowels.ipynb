# 這題要將字串後面的母音移除 (\#String )
原題目連結: https://leetcode.com/problems/trim-trailing-vowels/description/  

You are given a string s that consists of lowercase English letters.  
Return the string obtained by removing all trailing vowels from s.  
The vowels consist of the characters `'a'`, `'e'`, `'i'`, `'o'`, and `'u'`.

範例:  
Ex1:  
Input: s = "idea"  
Output: "id"  
Explanation:  
Removing "id`ea`", we obtain the string "id".  

Ex2:  
Input: s = "day"  
Output: "day"  
Explanation:  
There are no trailing vowels in the string "day".  

Ex3:  
Input: s = "aeiou"  
Output: ""  
Explanation:  
Removing "`aeiou`", we obtain the string "".  

* 解題想法:  
用一個變數來判斷是否需要移除，接著從字尾開始檢查，如果出現子音則更新變數，並將字母加到字串上，最後回傳字串

```python
class Solution:
    def trimTrailingVowels(self, s: str) -> str:
        res = ''
        is_vowel = True
        for i in range(len(s)-1, -1, -1):
            if is_vowel and s[i] not in ['a','e','i','o','u']:
                is_vowel = False
            if not is_vowel:
                res = s[i] + res
        return res
```
