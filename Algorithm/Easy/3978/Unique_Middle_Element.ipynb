# 本題要找出中位數是否為唯一值 (\#Array )
原題目連結: https://leetcode.com/problems/unique-middle-element/description/  

You are given an integer array nums of odd length n.  
Return `true` if the middle element of nums appears exactly once in the array. Otherwise return `false`.  

範例:  
Ex1:  
Input: nums = [1,2,3]  
Output: true  
Explanation:  
The middle element of nums is 2, which appears exactly once.  
Thus, the answer is `true`.  

Ex2:  
Input: nums = [1,2,2]  
Output: false  
Explanation:  
The middle element of nums is 2, which appears twice.  
Thus, the answer is `false`.

* 解題想法:   
首先將中位數的數值找出，並用一個變數來統計中位數出現的次數，最後回傳中位數是否出現超過一次

```python
class Solution:
    def isMiddleElementUnique(self, nums: list[int]) -> bool:
        m = nums[len(nums) // 2]
        count = 0
        for i in nums:
            if i == m:
                count += 1
        return count == 1
```
