# 本題要找出是否只出現一對連續的bit (\#BitManipulation )
原題目連結: https://leetcode.com/problems/exactly-one-consecutive-set-bits-pair/description/  

You are given an integer n.  
Return true if its binary representation contains exactly one pair of consecutive set bits, and false otherwise.  

範例:  
Ex1:  
Input: nums = 6  
Output: true  
Explanation:  
- Binary representation of 6 is 110.  
- There is exactly one pair of consecutive set bits ("11"). Thus, the answer is true.

Ex2:  
Input: nums = 5  
Output: false  
Explanation:  
- Binary representation of 5 is 101.
- There are no consecutive set bits. Thus, the answer is false.

* 解題想法:  
首先找出所有連續出現的bit，並檢查是否只有剛好一對相鄰的bit，如果有則回傳True，否則回傳False

```python
class Solution:
    def consecutiveSetBits(self, n: int) -> bool:
        b = bin(n)[2:]
        t = {}
        prev = ''
        for i in b:
            if i == '1':
                prev += i
            else:
                if prev and len(prev) > 1:
                    if len(prev) not in t.keys():
                       t[len(prev)] = 0
                    t[len(prev)] += 1 
                prev = ''
        if prev and len(prev) > 1:
            if len(prev) not in t.keys() :
                t[len(prev)] = 0
            t[len(prev)] += 1
        if t:
            if len(t) > 1:
                return False
            elif len(t) == 1:
                if 2 in t.keys() and t[2] == 1:
                    return True
                else:
                    return False
        return False
```
